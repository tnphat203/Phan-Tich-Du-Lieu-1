```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```


1.	Tạo 1 mảng dữ liệu ngẫu nhiên gồm 100 mẫu và 2 đặc trưng




```python
data = np.random.rand(100, 2)
data
```

2.	Scatter, quan sát và quyết định số cụm 

```python
plt.scatter(data[:, 0], data[:, 1])
plt.title('Scatter plot of data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

Chia làm 3 cụm:
* Cụm 1: Nằm ở góc dưới bên trái của sơ đồ, cụm này bao gồm các điểm dữ liệu có giá trị thấp cho cả hai đặc trưng.
* Cụm 2: Nằm ở góc trên bên phải của sơ đồ, cụm này bao gồm các điểm dữ liệu có giá trị cao cho cả hai đặc trưng.
* Cụm 3: Nằm ở trung tâm của sơ đồ, cụm này bao gồm các điểm dữ liệu có giá trị trung bình cho cả hai đặc trưng.

3.	Dùng kmean để phân cụm với số cụm như câu 2

```python
# Đặt số lượng cụm là 3
n_clusters = 3

# Khởi tạo và fit mô hình KMeans với số lượng cụm đã đặt và dữ liệu đầu vào
kmeans = KMeans(n_clusters=n_clusters, random_state=0).fit(data)

# In ra các điểm trung tâm của các cụm sau khi đã fit mô hình
print(kmeans.cluster_centers_)
```

4.	Trực quan với mỗi cụm mỗi màu khác nhau, đánh dấu các điểm trung tâm của các cụm

```python
# Lấy nhãn của các điểm dữ liệu sau khi đã được phân cụm
labels = kmeans.labels_

# Lấy các điểm trung tâm của các cụm
centers = kmeans.cluster_centers_

# Tạo biểu đồ scatter với mỗi cụm một màu
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis')

# Đánh dấu các điểm trung tâm của các cụm với màu đỏ và hình dạng 'X'
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='X')

plt.title('Scatter plot of data with clusters')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```


5.	Tạo ngẫu nhiên 1 mảng dữ liệu gồm 5 mẫu và 2 đặng trưng, dùng kmean và dữ liệu đã được huấn luyện trên để xác định mỗi mẫu thuộc cụm nào

```python
# Tạo ngẫu nhiên một mảng dữ liệu gồm 5 mẫu và 2 đặc trưng
new_data = np.random.rand(5, 2)

# Sử dụng mô hình KMeans đã được huấn luyện để xác định mỗi mẫu thuộc cụm nào
new_labels = kmeans.predict(new_data)

# In ra nhãn của mỗi mẫu
print(new_labels)
```
